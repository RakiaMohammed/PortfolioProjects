```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,6)
pd.options.mode.chained_assignment = None


df= pd.read_csv('/Users/rakia/Downloads/movies.csv')
```

```python
df
```

```python
# Missing Data Percentage for each Column:

for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print("{}-{}%".format(col,round(pct_missing*100)))
    
```

```python
# Find Datatypes of columns:

print(df.dtypes)
```

```python
# To find outliers of a particular column:

df.boxplot(column=['gross'])
```

```python
# Drop duplicates:

df.drop_duplicates()

#df['company'].drop_duplicates().sort_values(ascending=False): Drops duplicates from a particular column
#pd.set_option=('display.max_rows', None) : Displays all rows of df
```

```python
# Order data based on a column:

df.sort_values(by=['gross'], inplace=False, ascending= False)
```

```python
#Scatterplot for Budget vs Gross:

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Check Correlation between budget and gross:

sns.regplot(x='gross', y='budget', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Check Correlation between score and gross:

sns.regplot(x="score", y="gross", data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
#Correlation matrix between all numeric columns:

df.corr(method='pearson')  #methods: pearson, kendall, spearman
```

```python
df.corr(method='spearman')
```

```python
# Correlation matrix using heatmap:

correlation_matrix= df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()

#Conclusion1: High correlation between Budget and Gross.
#Conclusion2: Low correlation between Score and Gross.
```

```python
# Changing Non-numeric columns to Numeric columns:

df_numerized = df
for col in df_numerized.columns:
    if (df_numerized[col].dtype=='object'):
        df_numerized[col]= df_numerized[col].astype('category')
        df_numerized[col]=df_numerized[col].cat.codes
        
df_numerized




    
```

```python
df_numerized.corr(method='pearson')
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Movies")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
plt.show()
```

```python
correlation_mat = df_numerized.corr(method='pearson')

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)

```
